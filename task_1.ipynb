```python
def sum_distance(from_num, to_num):
    #from_num: начальное число
    #to_num: конечное число
   
    # Если первое число больше второго, меняем их местами
    if from_num > to_num:
        from_num, to_num = to_num, from_num
    
    # Суммируем все числа от from_num до to_num включительно
    total = sum(range(from_num, to_num + 1))
    
    return total

print(sum_distance(1, 10))
print(sum_distance(10, 1))
print(sum_distance(10, 10))
print(sum_distance(-3, 3))
print(sum_distance(-3, 10))
print(sum_distance(1, 100))
```
